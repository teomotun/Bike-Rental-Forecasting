```python
# Import 'mtcars' data from ggplot
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ggplot import mtcars
```

```python
mtcars.head()
```

```python
# Get the header data for 'mtcars' and store it in variable. 
# Display the header data using print() command.
header_data = list(mtcars.columns)
print(header_data)

# Get a quick description of the data
mtcars.info()
```

```python
# Calculate the mean of each of the columns of mtcars.
mtcars.mean()
```

```python
# Get the mean and median of each row (car) 
mtcarsMeanCarWise = mtcars.mean(axis=1)
medianCarsFeatures = mtcars.median(axis=1)
```

```python
# Plot of histogram to get the feel of type of data
mtcars.hist(bins=10, figsize=(20,15))
plt.show
```

```python
# Plot the density curves for each column of mtcars using 'for' loop and plot command
for cols in mtcars.columns: 
    if (cols != "name"): 
        mtcars[cols].plot(kind="density",figsize=(2,2)) 
        plt.title(cols) 
        plt.show()
```

```python
# Plot scatter matrix
from pandas.plotting import scatter_matrix
scatter_matrix(mtcars[["mpg","disp","hp","wt","qsec"]], figsize=(12, 8))
plt.show()
```

```python
# Calc skewness value of each column
skewValue = mtcars.skew(axis=0)
print(skewValue)
```

```python
# Get summary statistics
mtcars.describe()
```

```python
# Get interquantile spread of the data
mtcars.quantile()
```

```python
# Box plot for 'mpg' column of mtcars.
mtcars.boxplot("mpg")
plt.show()
```

```python
print(mtcars["mpg"].mean())
print(mtcars["mpg"].median())
print(mtcars["mpg"].std())
```

```python
# Seperate automatic cars and manual cars into two groups
mtCarsAutomatic = mtcars.loc[mtcars["am"]==1]
mtCarsManual = mtcars.loc[mtcars["am"]==0]
```

```python
# Plot the box plot for 'mpg' (miles per gallons) for each of the subsets of cars grouped by transmission
mtCarsAutomatic.boxplot("mpg")
plt.title("Automatic")
plt.show()
mtCarsManual.boxplot("mpg")
plt.title("Manual")
plt.show()
```

## From the box-plot, it can be inferred that automatic cars have higher mpg than manual cars.
## We can also make similar box-plots using carburetors, gears and engine shape.
